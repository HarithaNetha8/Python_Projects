```python
NOW IMNCREATIM A ANALYSIS FOR RETAIL STORE TO UNDERSTAND TE PYTON LIBRARIES AND ANALYSIS T EREALTIME DATASET
OBJECTIVES

-> revenue trends
-> best sellin products
-> top cutomers
-> montly sales rowt
-> business insits
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('retail_store_dataset.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df['revenue']=df['Quantity']*df['Final_Unit_Price']
df
```

```python
df.columns
```

```python
df['Order_Date']=pd.to_datetime(df['Order_Date'])
```

```python
df['Month']=df['Order_Date'].dt.month
```

```python
df['year']=df['Order_Date'].dt.year
```

```python
df
```

```python
monthly_Revenue=df.groupby('Month')['revenue'].sum()
print(monthly_Revenue)
```

```python
7) top customers
```

```python
top_customer= df.groupby('Order_ID')['revenue'].sum().sort_values(ascending=False).head(10)
print(top_customer)
```

```python
8) best selling products - top 5 procucts
```

```python
best_sp=df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10)
print(best_sp)
```

```python
9)revenue by category
```

```python
r_c=df.groupby('Category')['revenue'].sum()
print(r_c)
```

```python
10)visulization
```

```python
# barchar - top products
best_sp.plot(kind='bar')
plt.title('top 10 products')
plt.show()
```

```python
# line cart - montly revenue
monthly_Revenue.plot(kind='line')
plt.title('montly sales revenue')
plt.show()
```

```python
r_c.plot(kind='pie',autopct='%1.1f')
plt.title('revenue by category')
plt.show()
```

```python
#profit by reion
p_r=df.groupby('Region')['Profit'].sum()
print(p_r)
```

```python
p_r.plot(kind='bar')
plt.title('profit by region')
plt.show()
```

```python
#payment method analsysis

payment=df['Payment_Method'].value_counts()
print(payment)
```

```python
payment.plot(kind='pie',autopct='%1.1f', explode=[0.1,0,0,0])
plt.title('payment method analsysis')
plt.show()
```

**insights**

1) sports category generated the highest revenue among all categories.

2) Most customers prefer Wallet when comparing to UPI,Card and Cash.

3) Sales peak during mid-year months due to seasonal demand.

4) west Region have more number of profit
