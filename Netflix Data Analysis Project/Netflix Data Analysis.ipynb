**Netflix Data Analysis (Data Analyst Project)**

```python
#**importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading dataset

df=pd.read_csv('netflix.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull() 
```

```python
#Data Cleaning
```

```python
df.isnull().sum()
```

```python
# Fill missing values:
# df["country"].fillna("Unknown", inplace=True)
# df["director"].fillna("Unknown", inplace=True)
#you can fill na values with different values or mean meadian mode for numerical values or 0...
```

```python
df["cast"].fillna("Unknown",inplace=True)
df["date_added"].fillna("Not Available",inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["duration"].fillna(df["duration"].mode()[0], inplace=True)
```

```python
df["rating"].fillna(df["rating"].mode()[0], inplace=True)
df
```

```python
# Movies vs TV Shows

ct=df["type"].value_counts()
print(ct)
```

```python
# Visualization

ct.plot(kind="bar")
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python

ct.plot(kind="pie")
plt.title('Movies vs TV Shows on Netflix in pie representation')
plt.show()
```

```python
# Content by Country

t_c=df['country'].value_counts().head(5)
t_c
```

```python
t_c.plot(kind='bar')
plt.title('Top 10 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Shows')
plt.show()
```

```python
# Content Growth Over Years

year_content=df['release_year'].value_counts().sort_index()
print(year_content)
```

```python
year_content.plot(kind='line',color='red')
plt.title('Netflix Content Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Shows')
plt.show()
```

```python
# Most Popular Genres

genres = df["listed_in"].str.split(",", expand=True).stack()
# top_genres = genres.value_counts().head(10)

print(genres)
```

```python
top_genres = genres.value_counts().head(10)
print(top_genres)
```

```python
top_genres.plot(kind="bar")

plt.title("Top Genres on Netflix")
plt.show()
```
